```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
# define file path and sheet names
file_path = r"c:\Users\jari5003\Desktop\Selen  P&G\MAIN_Penetration and Value Drivers Template_raw data for Divyaa & Rinu.xlsx"
sheet1 = "Penetration Trends Template"
sheet2 = "KPIs - EU5 template"

# load both sheets skipping raw1 and assigning multi-row headers
# df1: skip first row then next 4 rows as headers
# df2: skip first row then next 2 rows as headers

df1 = pd.read_excel(
    file_path,
    sheet_name=sheet1,
    skiprows=1,
    header=[0,1,2,3]
)

df2 = pd.read_excel(
    file_path,
    sheet_name=sheet2,
    skiprows=1,
    header=[0,1]
)

# show some info
print("df1 shape:", df1.shape)
print("df1 columns:\n", df1.columns)
print("df2 shape:", df2.shape)
print("df2 columns:\n", df2.columns)
```

```python
# Replace Unnamed columns in the 3rd header level (level index 2) after column 7 with 'Total Europe'
if isinstance(df1.columns, pd.MultiIndex):
    cols = list(df1.columns)
    new_cols = []
    for i, col in enumerate(cols):
        col_list = list(col)
        if i >= 7 and len(col_list) > 2 and isinstance(col_list[2], str) and col_list[2].startswith('Unnamed'):
            col_list[2] = 'Total Europe'
        new_cols.append(tuple(col_list))
    df1.columns = pd.MultiIndex.from_tuples(new_cols)
else:
    cols = df1.columns.tolist()
    for i in range(7, len(cols)):
        if isinstance(cols[i], str) and cols[i].startswith('Unnamed'):
            cols[i] = 'Total Europe'
    df1.columns = cols
# display updated dataframe
df1
```

```python
# previously we modified headers; now drop the first two header levels for clarity
# this effectively removes the top two rows of the MultiIndex
if isinstance(df1.columns, pd.MultiIndex) and df1.columns.nlevels > 2:
    df1.columns = df1.columns.droplevel([0,1])

print("Columns after dropping first two levels:\n", df1.columns[:10])
```

```python
# alternative transformation using melt, treating first 7 columns as metadata
# use df_base from earlier steps
try:
    df_base = df1_trimmed
except NameError:
    df_base = df1

# determine metadata and data columns by position
metadata_cols = list(df_base.columns[:7])
# we'll slice by position for data rather than by labels

data_cols_raw = list(df_base.columns[7:])

print("Metadata cols:", metadata_cols)
print("First few raw data cols:", data_cols_raw[:5])

# modify header2 (metric) names for data columns starting from original position 26 onwards
# original col index = 7 + idx in data_cols_raw
renamed_data = []
for idx, col in enumerate(data_cols_raw):
    c0, c1 = col
    if (7 + idx) >= 25:  # zero-based original
        c1 = f"{c1} vs YA"
    renamed_data.append((c0, c1))

# slice the data portion of df_base and then rename its columns

df_data = df_base.iloc[:, 7:].copy()
df_data.columns = pd.MultiIndex.from_tuples(renamed_data)

# also extract df_meta for later use

df_meta = df_base.iloc[:, :7].copy()

# flatten df_data columns
flat_cols = [f"{c0}|||{c1}" for c0, c1 in renamed_data]
df_flat = df_data.copy()
df_flat.columns = flat_cols

# add meta columns to flat
for col in metadata_cols:
    colname = col[1]
    df_flat[f"meta|||{colname}"] = df_meta[col].squeeze().values

# preserve row identifier
df_flat = df_flat.reset_index(drop=True)
df_flat['row_id'] = df_flat.index

# melt on metric columns
id_vars = ['row_id'] + [f"meta|||{col[1]}" for col in metadata_cols]
df_melt = df_flat.melt(id_vars=id_vars, var_name='combined', value_name='value')

# split combined
split_cols = df_melt['combined'].str.split('\\|\\|\\|', expand=True)
df_melt['country'] = split_cols[0]
df_melt['metric'] = split_cols[1]

# pivot back
df_final = df_melt.pivot_table(
    index=id_vars + ['country'],
    columns='metric',
    values='value',
    aggfunc='first'
).reset_index()

# clean metadata column names
for col in metadata_cols:
    df_final = df_final.rename(columns={f"meta|||{col[1]}": col[1]})

print("Transformed final shape:", df_final.shape)
print(df_final.head())
```

# pg risk
- df_final[df_final['BRAND OWNER']=='PROCTER & GAMBLE'] if the Latest 52 vs YA is positive but if Latest 26 vs YA or Latest 13 vs YA is smaller than Latest 52 vs YA then it is declining growt

```python
# define a helper that applies the two rules only for P&G brands
def calculate_pg_risk(row):
    if row['BRAND OWNER'] != 'PROCTER & GAMBLE':
        return 'No'                       # keep others as “No”
    l52 = row['Latest 52 vs YA']
    l26 = row['Latest 26 vs YA']
    l13 = row['Latest 13 vs YA']
    # declining‑growth
    if l52 > 0 and (l26 < l52 or l13 < l52):
        return 'Declining growth'
    # declining‑deceleration
    if l52 < 0 and (l26 > l52 or l13 > l52):
        return 'Declining Deceleration'
    return 'No'

# assign the column
df_final['P&G risk'] = df_final.apply(calculate_pg_risk, axis=1)

# quick check
print(df_final['P&G risk'].value_counts())
print(df_final[df_final['BRAND OWNER']=='PROCTER & GAMBLE'][['BRAND','country',
      'Latest 52 vs YA','Latest 26 vs YA','Latest 13 vs YA','P&G risk']].head())
```

```python
# RECHECK: Full verification of P&G risk logic starting with P&G data
print("="*80)
print("FULL RECHECK: LOGIC VERIFICATION")
print("="*80)

# Step 1: Filter for PROCTER & GAMBLE brands
pg_brands = df_final[df_final['BRAND OWNER']=='PROCTER & GAMBLE'].copy()
print(f"\n✓ Total P&G rows in df_final: {len(pg_brands)}")
print(f"✓ Total rows in df_final: {len(df_final)}")

# Step 2: Check P&G risk column exists
if 'P&G risk' in df_final.columns:
    print(f"✓ P&G risk column exists in df_final")
else:
    print("✗ ERROR: P&G risk column NOT found in df_final")

# Step 3: Verify logic for each risk category
print("\n" + "="*80)
print("RISK CATEGORY VERIFICATION")
print("="*80)

# Declining growth: Latest 52 > 0 AND (Latest 26 < Latest 52 OR Latest 13 < Latest 52)
declining_growth = pg_brands[
    (pg_brands['Latest 52 vs YA'] > 0) & 
    ((pg_brands['Latest 26 vs YA'] < pg_brands['Latest 52 vs YA']) | 
     (pg_brands['Latest 13 vs YA'] < pg_brands['Latest 52 vs YA']))
]

declining_growth_marked = pg_brands[pg_brands['P&G risk'] == 'Declining growth']

print(f"\n1. DECLINING GROWTH:")
print(f"   Expected logic count: {len(declining_growth)}")
print(f"   Marked as 'Declining growth': {len(declining_growth_marked)}")
print(f"   Match: {'✓ YES' if len(declining_growth) == len(declining_growth_marked) else '✗ NO'}")

# Declining Deceleration: Latest 52 < 0 AND (Latest 26 > Latest 52 OR Latest 13 > Latest 52)
declining_deceleration = pg_brands[
    (pg_brands['Latest 52 vs YA'] < 0) & 
    ((pg_brands['Latest 26 vs YA'] > pg_brands['Latest 52 vs YA']) | 
     (pg_brands['Latest 13 vs YA'] > pg_brands['Latest 52 vs YA']))
]

declining_deceleration_marked = pg_brands[pg_brands['P&G risk'] == 'Declining Deceleration']

print(f"\n2. DECLINING DECELERATION:")
print(f"   Expected logic count: {len(declining_deceleration)}")
print(f"   Marked as 'Declining Deceleration': {len(declining_deceleration_marked)}")
print(f"   Match: {'✓ YES' if len(declining_deceleration) == len(declining_deceleration_marked) else '✗ NO'}")

# No risk: Everything else (including non-P&G)
no_risk_pg = pg_brands[pg_brands['P&G risk'] == 'No']
print(f"\n3. NO RISK (P&G brands only):")
print(f"   Count: {len(no_risk_pg)}")

# Step 4: Overall distribution
print("\n" + "="*80)
print("OVERALL P&G RISK DISTRIBUTION")
print("="*80)
print(df_final[df_final['BRAND OWNER']=='PROCTER & GAMBLE']['P&G risk'].value_counts())

print("\n" + "="*80)
print("NON-P&G BRANDS - SHOULD ALL BE 'No'")
print("="*80)
non_pg = df_final[df_final['BRAND OWNER']!='PROCTER & GAMBLE']
non_pg_risk_dist = non_pg['P&G risk'].value_counts()
print(non_pg_risk_dist)
if len(non_pg_risk_dist) == 1 and non_pg_risk_dist.index[0] == 'No':
    print("✓ All non-P&G brands correctly marked as 'No'")
else:
    print("✗ ERROR: Some non-P&G brands have risk values other than 'No'")

# Step 5: Sample verification rows
print("\n" + "="*80)
print("SAMPLE ROWS - DETAILED VERIFICATION")
print("="*80)

print("\nSample DECLINING GROWTH rows:")
declining_growth_sample = declining_growth_marked[['BRAND', 'country', 'Latest 52 vs YA', 'Latest 26 vs YA', 'Latest 13 vs YA', 'P&G risk']].head(5)
print(declining_growth_sample)
print("Logic check (for each row, Latest 52 > 0 and min(26,13) < 52):")
for idx, row in declining_growth_sample.iterrows():
    l52 = row['Latest 52 vs YA']
    l26 = row['Latest 26 vs YA']
    l13 = row['Latest 13 vs YA']
    print(f"  {row['BRAND']}: 52={l52:.3f} (>0?{l52>0}), 26={l26:.3f}(<52?{l26<l52}), 13={l13:.3f}(<52?{l13<l52})")

print("\nSample DECLINING DECELERATION rows:")
declining_deceleration_sample = declining_deceleration_marked[['BRAND', 'country', 'Latest 52 vs YA', 'Latest 26 vs YA', 'Latest 13 vs YA', 'P&G risk']].head(5)
print(declining_deceleration_sample)
print("Logic check (for each row, Latest 52 < 0 and max(26,13) > 52):")
for idx, row in declining_deceleration_sample.iterrows():
    l52 = row['Latest 52 vs YA']
    l26 = row['Latest 26 vs YA']
    l13 = row['Latest 13 vs YA']
    print(f"  {row['BRAND']}: 52={l52:.3f} (<0?{l52<0}), 26={l26:.3f}(>52?{l26>l52}), 13={l13:.3f}(>52?{l13>l52})")

print("\n" + "="*80)
print("VERIFICATION COMPLETE")
print("="*80)
```
